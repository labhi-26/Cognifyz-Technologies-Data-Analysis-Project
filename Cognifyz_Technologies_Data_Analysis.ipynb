## Task: Price Range Distribution
### Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants

```python
price_range = df['Price range']
```

```python
price_range.value_counts()
```

```python
total_restaurants = len(price_range)
```

```python
total_restaurants
```

```python
percentage = (price_range.value_counts() / total_restaurants) * 100
```

```python
percentage
```

```python
import matplotlib.pyplot as plt

price_range_counts = {}  # store the count of each unique price range category

for price in price_range:
    if price in price_range_counts:
        price_range_counts[price] += 1
    else:
        price_range_counts[price] = 1

sorted_price_range_counts = dict(sorted(price_range_counts.items()))

total_restaurants = len(price_range)

percentage = {price: (count / total_restaurants) * 100 for price, count in sorted_price_range_counts.items()}

plt.figure(figsize=(8, 6))
bars = plt.bar(percentage.keys(), percentage.values(), color='lightgreen')

plt.xlabel('Price Range')
plt.ylabel('Percentage of Restaurants (%)')
plt.title('Distribution of Price Ranges Among Restaurants')

plt.xticks(list(percentage.keys()))

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.ylim(0, 100)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}%', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

### Calculate the percentage of restaurants in each price range category

```python
for price, count in percentage.items():
    print(f"Percentage of restaurants in Price Range category {price}: {count:.2f}%")
```

### Identify any patterns or clusters of restaurants in specific areas

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = pd.read_excel('/content/drive/MyDrive/Dataset.excel 1.txt.xlsx')

X = data[['Latitude', 'Longitude']]

# Choose the number of clusters
n_clusters = 5

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
data['Cluster'] = kmeans.fit_predict(X)

# Visualize the clusters
plt.figure(figsize=(10, 8))
for cluster_label in range(n_clusters):
    cluster_data = data[data['Cluster'] == cluster_label]
    plt.scatter(cluster_data['Longitude'], cluster_data['Latitude'], label=f'Cluster {cluster_label}')

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Restaurant Clusters')
plt.legend()
plt.grid(True)
plt.show()
```

```python


# 2. Cluster Characteristics
cluster_counts = data['Cluster'].value_counts()
print("Cluster Counts:")
print(cluster_counts)

# Example: Print the first few rows of each cluster to see the characteristics
for cluster_label in range(n_clusters):
    cluster_data = data[data['Cluster'] == cluster_label]
    print(f"\nCluster {cluster_label} Characteristics:")
    print(cluster_data.head())
```

### Identify if there are any restaurant chains present in the dataset

```python
restaurant_counts = df['Restaurant Name'].value_counts()
```

```python
restaurant_chains = restaurant_counts[restaurant_counts > 1]
```

```python
if len(restaurant_chains) > 0:
    print("Restaurant Chains Present in the Dataset:")
    print(restaurant_chains)
else:
    print("No restaurant chains found in the dataset.")
```

### Analyze the ratings and popularity of different restaurant chains

```python
restaurant_counts = df['Restaurant Name'].value_counts()
restaurant_chains = restaurant_counts[restaurant_counts > 1].index.tolist()

chain_ratings = {}
for chain in restaurant_chains:
    chain_df = df[df['Restaurant Name'] == chain]
    mean_rating = chain_df['Aggregate rating'].mean()
    total_ratings = chain_df['Aggregate rating'].count()
    num_locations = chain_df['Restaurant Name'].nunique()
    total_votes = chain_df['Votes'].sum()
    chain_ratings[chain] = {'Mean Rating': mean_rating, 'Total Ratings': total_ratings, 'Number of Locations': num_locations, 'Total Votes': total_votes}

chain_ratings_df = pd.DataFrame.from_dict(chain_ratings, orient='index')

chain_ratings_df = chain_ratings_df.sort_values(by='Total Ratings', ascending=False)

print("Analysis of Ratings and Popularity for Different Restaurant Chains:")
print(chain_ratings_df)
```
